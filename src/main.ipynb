```python
from numpy import *
import matplotlib.pyplot as plt
```

```python
# example
l = arange(1, 20, 0.1)  # cm
g = 1000  # cm / s^2
m = 100  # g
_h = pow(10, -27) # erg * s = g * cm ^ 2 / s^ 2
```

```python
keisu = sqrt(l / g)
log_ = log((pi / 180) * sqrt(2 * m * sqrt(g * (l ** 3)) / _h))
t = keisu * log_  # s
# print(t)
```

```python
plt.plot(l, t, label="time(s)")
# 凡例の表示
plt.legend()
plt.xlabel('drill length')
plt.ylabel('time(s)')
# プロット表示(設定の反映)
plt.show()
```
